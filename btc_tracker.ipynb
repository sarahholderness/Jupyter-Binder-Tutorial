```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=BTC&market=USD&apikey=KN9JDGZ91NBNFGET'

r = requests.get(url)
data = r.json()
# Get the inner dictionary with the daily data
df2 = pd.DataFrame.from_dict(data['Time Series (Digital Currency Daily)'], orient='index')
# Rename the first column (which is actually the index) to Date
df2.index.names = ['Date']
# Make the index a Datetime
df2.index = pd.to_datetime(df2.index)
# Make the closing value a float (decimal number)
df2['4b. close (USD)'] = pd.to_numeric(df2['4b. close (USD)'], downcast="float")
# Reverse the order of the dates
df3 = df2.iloc[::-1]
# Plot the graph
df3['4b. close (USD)'].plot()
plt.tight_layout()
plt.title('Daily close value for bitcoin (BTC)')
plt.grid()
plt.show()
```
